# 3. Low Level Data Understanding :

a. Find count of unique values in location column.

```python
count_uniq = df['location'].nunique()
print("Count of uniques values in location:",count_uniq)
```

b. Find which continent has maximum frequency using values 
counts.

```python
max_freq = df['continent'].value_counts().max()
print("Maximum frequency:",max_freq)
```

c. Find maximum & mean value in 'total_cases'.

```python
maxi = df['total_cases'].max()
mean = df['total_cases'].mean()
print(f"Maximum Value: {maxi}")
print(f"Mean Value: {mean}")
```

d. Find 25%,50% & 75% quartile value in 'total_deaths'.

```python
q1 = df['total_deaths'].quantile(0.25)
q2 = df['total_deaths'].quantile(0.50)
q3 = df['total_deaths'].quantile(0.75)

print(f"25% quartile value:{q1}")
print(f"50% quartile value:{q2}")
print(f"75% quartile value:{q3}")
```

 e. Find which continent has maximum 'human_development_index'.

```python
max_human_id = df.groupby('continent')['human_development_index'].max()
max_con_hdi =max_human_id[max_human_id == max_human_id.max()].index[0]
print(f"The continent which has maximum human development index: {max_con_hdi}")
```

f. Find which continent has minimum 'gdp_per_capita'.

```python
mini_gdp = df.groupby('continent')['gdp_per_capita'].min()
mini_con_gdp=mini_gdp[mini_gdp == mini_gdp.min()].index[0]
print(f"The continent which has minimum gdp per capita: {mini_con_gdp}")
```

# 5. Data Cleaning

# 7. Data Aggregation:

 a. Find max value in all columns using groupby function on 
'continent'
 column

```python
maximum = df.groupby('continent').max().reset_index()
maximum
```

 b. Store the result in a new dataframe named 'df_groupby'.

```python
df_groupby = df.groupby('continent').max().reset_index()
df_groupby
```

# 8. Feature Engineering :

a. Create a new feature 'total_deaths_to_total_cases' by ratio of
 'total_deaths' column to 'total_cases'


```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
df_groupby
```
